```python
import pandas as pd
import matplotlib.pyplot as mat
```

```python


df = pd.read_csv("/content/drive/MyDrive/Python/rounds_in_games.csv")
df_games = pd.read_csv("/content/drive/MyDrive/Python/games.csv")
df_words = pd.read_csv("/content/drive/MyDrive/Python/words.csv", header=None, names=["Words"])
```

## **WORDS**
## Distribución de preguntas por número de letras.
De las palabras del archivo words.csv

```python
#Extraemos la longitud de todas las palabras, a través del nombre de cabecera
#dado en la lectura del archivo

#Añadida la columna Length, donde se mostrará la longitud de cada palabra.

df_words["Length"] = df_words["Words"].apply(len)
df_words
```

De las palabras del archivo rounds_in_games.csv

```python
words = df["Word"]
```

```python
#Añadida columna con la longitud de las palabras que se han jugado en las rondas
df["Length"] = df["Word"].apply(len)
df
```

```python
#Agrupamos por longitud e intentos de usuario.
#En una nueva columna mostramos el número de palabras coincidentes por longitud e intentos de usuario
lengthAndAttempts = df.groupby(["Length", "User Tries"]).size().reset_index(name="How Many")
lengthAndAttempts
```

```python
#Array con las longitudes únicas
uniques = lengthAndAttempts["Length"].unique()
uniques
```

```python
#Recorremos el array.
#A partir de ahí se crearán los gráficos
#Para facilitar la visualización le damos título, etiquetas por cada eje de la gráfica, etc...
#
for u in uniques:
  results = lengthAndAttempts[lengthAndAttempts["Length"] == u]
  mat.bar(results["User Tries"], results["How Many"],zorder=2)
  mat.title(f"Intentos con palabras de {u} letras")
  mat.xlabel("Intentos")
  mat.ylabel("Número de palabras")
  mat.xticks(results["User Tries"])
  mat.yticks(results["How Many"]) #Los ticks eliminan los valores no usados de la gráfica
  mat.grid(True, axis="y", linestyle="--", linewidth=1, zorder=0)
  mat.show()
```

##Top 5 de usuarios que más rondas han ganado por partida

```python
#Extraemos las rondas que acabaron en victoria
victories = df[df["Victory"]==True]

#Creamos el ranking con aquellos que han tenido más victorias.
ranking = victories["Username"].value_counts().head(5).reset_index(name="Won_Rounds")
ranking
```

## Top 5 de usuarios que más rondas han ganado por partida (Primer intento)

```python


victories = df[df["Victory"]==True]
rondas = victories.groupby(["Username"]).size().reset_index(name="Rondas Ganadas").head(5)
top5 = rondas.sort_values(by="Rondas Ganadas", ascending=False)
top5
```

## **GAMES**
## Top 5 de usuarios con más puntuación final


```python


top5 = df_games.groupby("Username")["Final Score"].sum().nlargest(5).reset_index()
top5.columns = ["Username","Highest Scores"]
top5
```

## Top 5 de usuarios con más puntuación media (media de la puntuación de sus partidas)

```python
top5Avg = df_games.groupby("Username")["Final Score"].mean().nlargest(5).reset_index()
top5Avg.columns = ["Username", "Average_Score"]
top5Avg
```

```python
mat.hist(df_games["Final Score"],zorder=2)
mat.title('Distribución de puntuaciones')
mat.xlabel('Puntuación')
mat.ylabel('Numero de partidas')
mat.grid(True, axis='y', linestyle='--', linewidth=1, zorder=0)#zorder = capa para que se vea por debajo de las barras
mat.xticks([0, 1, 2, 3])  # Establece los valores de las posibles puntuaciones finales

mat.show()
```
